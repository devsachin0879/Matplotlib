#### Axes labels,legend and grid

```python
days = [1,2,3,4,5,6,7]
max_t = [50,51,52,48,47,49,46]
min_t = [43,42,40,44,33,35,37]
avg_t = [45,48,48,46,40,42,41]
```

```python
plt.xlabel("Days")
plt.ylabel("Temperature")
plt.title("Weather")

plt.plot(days,max_t,label = "Max")
plt.plot(days,min_t,label = "Min")
plt.plot(days,avg_t,label = "Avg")

plt.legend()
```

```python
plt.xlabel("Days")
plt.ylabel("Temperature")
plt.title("Weather")

plt.plot(days,max_t,label = "Max")
plt.plot(days,min_t,label = "Min")
plt.plot(days,avg_t,label = "Avg")

plt.legend(loc = "upper right")
```

```python
plt.xlabel("Days")
plt.ylabel("Temperature")
plt.title("Weather")

plt.plot(days,max_t,label = "Max")
plt.plot(days,min_t,label = "Min")
plt.plot(days,avg_t,label = "Avg")

plt.legend(loc = "best")
```

```python
plt.xlabel("Days")
plt.ylabel("Temperature")
plt.title("Weather")

plt.plot(days,max_t,label = "Max")
plt.plot(days,min_t,label = "Min")
plt.plot(days,avg_t,label = "Avg")

plt.legend(loc = "best",shadow = True,fontsize = "small")
```

```python
plt.xlabel("Days")
plt.ylabel("Temperature")
plt.title("Weather")

plt.plot(days,max_t,label = "Max")
plt.plot(days,min_t,label = "Min")
plt.plot(days,avg_t,label = "Avg")

plt.legend(loc = "best",shadow = True,fontsize = "small")
plt.grid()
```

#### Bar chart

```python
company = ['GOOGL','AMZN','MSFT','FB']
revenue = [90,136,89,27]
profit = [40,2,34,12]
```

```python
import numpy as np
xpos = np.arange(len(company))
plt.bar(company,revenue)
```

```python
plt.xticks(ypos,company)
plt.xlabel("Company Name")
plt.ylabel("revenue(bln)")
plt.title("US Tech Stocks")
plt.bar(xpos,revenue,label = "Revenue")
plt.legend()
```

```python
plt.xticks(ypos,company)
plt.xlabel("Company Name")
plt.ylabel("revenue(bln)")
plt.title("US Tech Stocks")
plt.bar(xpos,revenue,label = "Revenue")
plt.bar(xpos,profit,label = "Profit")
plt.legend()
```

```python
plt.xticks(ypos,company)
plt.xlabel("Company Name")
plt.ylabel("revenue(bln)")
plt.title("US Tech Stocks")
plt.bar(xpos-0.2,revenue,width = 0.4,label = "Revenue")
plt.bar(xpos+0.2,profit,width = 0.4,label = "Profit")
plt.legend()
```

```python
##horizontal bar chart
plt.yticks(ypos,company)
plt.xlabel("Company Name")
plt.ylabel("revenue(bln)")
plt.title("US Tech Stocks")
plt.barh(xpos-0.2,revenue,label = "Revenue")
plt.barh(xpos+0.2,profit,label = "Profit")
plt.legend()
```

#### Histograms

```python
blood_sugar = [113,85,90,150,149,88,93,115,135,80,77,82,129]
plt.hist(blood_sugar) ##by default num_of bins = 10
```

```python
blood_sugar = [113,85,90,150,149,88,93,115,135,80,77,82,129]
plt.hist(blood_sugar,bins = 3,rwidth = 0.87) ##r width is the relative width of a bar comapred to bin size
```

```python
blood_sugar = [113,85,90,150,149,88,93,115,135,80,77,82,129]
plt.hist(blood_sugar,bins = [80,100,125,150],rwidth = 0.87)
```

```python
blood_sugar = [113,85,90,150,149,88,93,115,135,80,77,82,129]
plt.hist(blood_sugar,bins = [80,100,125,150],rwidth = 0.87,color = 'g',histtype = 'step')
```

```python
blood_sugar_men = [113,85,90,150,149,88,93,115,135,80,77,82,129]
blood_sugarr_women = [67,98,89,120,133,150,84,69,89,79,120,112,100]
plt.xlabel('sugar range')
plt.ylabel('Total num of patients')
plt.title('Blood Sugar Analysis')
plt.hist([blood_sugar_men,blood_sugarr_women],bins = [80,100,125,150],rwidth = 0.87,color = ['green','orange'],
         label = ['men','women'])
plt.legend()
```

#### Pie Chart

```python
exp_values = [1400,600,300,410,250]
exp_labels = ['Home Rent','Food','Phone/Internet Bill','Car','Other Utilitties']
```

```python
plt.pie(exp_values,labels = exp_labels)
```

```python
plt.pie(exp_values,labels = exp_labels)
plt.show()
```

```python
plt.pie(exp_values,labels = exp_labels,radius = 1.5)
plt.show()
```

```python
plt.pie(exp_values,labels = exp_labels,radius = 1.5,autopct = '%0.2f%%')
plt.show()
```

```python
plt.pie(exp_values,labels = exp_labels,radius = 1.5,autopct = '%0.2f%%',shadow = True)
plt.show()
```

```python
plt.pie(exp_values,labels = exp_labels,radius = 1.5,autopct = '%0.2f%%',shadow = True,explode=[0,0.2,0.1,0,0] )
plt.show()
```

```python
plt.pie(exp_values,labels = exp_labels,radius = 1.5,autopct = '%0.2f%%',shadow = True,explode=[0,0.2,0.1,0,0],startangle = 45 )
plt.show()
```

```python
plt.pie(exp_values,labels = exp_labels,radius = 1.5,autopct = '%0.2f%%',shadow = True,explode=[0,0.1,0.1,0.1,0.15],startangle=180 )
plt.show()
```

#### save chart to file using savefig

```python
plt.pie(exp_values,labels = exp_labels,radius = 1.5,autopct = '%0.2f%%',shadow = True,explode=[0,0.1,0.1,0.1,0.15],startangle=180 )
plt.savefig("pie-chart.png",bbox_inches = 'tight',pad_inches = 2,tranparent = True)
```
